```python
# Importing Libraries
import ast
import numpy as np
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_us = df[(df['job_country']== 'United States')].dropna(subset=['salary_year_avg'])
```

```python
df_us
```

```python
job_titles= df_us['job_title_short'].value_counts().index[:6].to_list()
```

```python
df_top6jobs = df_us[df_us['job_title_short'].isin(job_titles)]
```

```python
job_order =df_top6jobs.groupby('job_title_short')['salary_year_avg'].agg('median').sort_values(ascending=False).index
```

```python
sns.boxenplot(data=df_top6jobs, x='salary_year_avg' , y ='job_title_short', order=job_order)
sns.set_theme(style='ticks')

plt.xlabel('yearly salary average')
plt.ylabel('')
plt.title('us salary distribution')
plt.xlim(0,600000)
plt.show()
```

Investigating median salary Vs skill for data analyst

```python
df_us_da = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States' )].copy()
```

```python
df_us_da = df_us_da.dropna(subset=['salary_year_avg'])
df_us_da = df_us_da.explode('job_skills')
```

```python
df_us_da[['salary_year_avg','job_skills']].head(5)
```

```python
top_pay = df_us_da.groupby('job_skills')['salary_year_avg'].agg(['count','median']).sort_values(by='median',ascending=False)

top_pay = top_pay.head(10)

top_pay
```

```python
top_skills =df_us_da.groupby('job_skills')['salary_year_avg'].agg(['count','median']).sort_values(by='count', ascending=False)

top_skills =top_skills.head(10).sort_values(by='median' , ascending=False)

top_skills
```

```python
fig, ax = plt.subplots(2,1)

sns.set_theme(style='ticks')

sns.barplot(data=top_pay , x='median' , y='job_skills' , hue='median' , palette='light:b' ,ax=ax[0])
ax[0].legend().remove()
ax[0].set_ylabel('')
ax[0].set_xlabel('')
ax[0].set_title('Top 10 highest paid skills in US')

sns.barplot(data= top_skills, x='median', y='job_skills', hue='median', palette='light:b' , ax=ax[1])
ax[1].legend().remove()
ax[1].set_ylabel('')
ax[1].set_xlabel('Median salary (US)')
ax[1].set_title('Top 10 highest in demand jobs in US')
ax[1].set_xlim(ax[0].get_xlim())


fig.tight_layout()
```
