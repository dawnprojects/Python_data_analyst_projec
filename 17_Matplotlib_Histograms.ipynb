```python
# Importing Libraries
import ast
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df
```

```python
df_us = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')] 
```

```python
df_us['salary_year_avg'].plot(kind='hist', bins= 40, edgecolor = 'black')
plt.xlim(0,200000)
plt.title('distribution of yearly salary of data analytics in USA')
plt.xlabel('yearly salry')
plt.ylabel('number of jobs  ')
```
