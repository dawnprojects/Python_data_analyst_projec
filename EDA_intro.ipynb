```python
# Importing Libraries
import ast
import numpy as np
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

Explotary Data Analysis 

```python
df_us = df[(df['job_title'] == 'Data analyst') & (df['job_country'] == 'United States')]
```

```python
df_plot = df_us['job_location'].value_counts().head(10).to_frame()
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count',y='job_location',hue='count', palette= 'dark:b_r',legend=False)
sns.despine()
plt.title('Number of jobs location at US location')
plt.xlabel('number of jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df_us['company_name'].value_counts().head(10).to_frame()
sns.barplot(data=df_plot, x='count' , y='company_name', hue='count', palette='dark:b_r', legend=False)
plt.title('Top companies in US')
plt.xlabel('number of jobs')
plt.ylabel('')
plt.show()
```
