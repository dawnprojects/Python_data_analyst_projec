```python
# Importing Libraries
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us=df[df['job_country'] == 'United States'].copy()
```

```python
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')
```

```python
df_us_pivot=df_us.pivot_table(index='job_posted_month' ,columns= 'job_title_short', aggfunc='size' )
df_us_pivot.reset_index(inplace=True)
df_us_pivot['months'] = pd.to_datetime(df_us_pivot['job_posted_month'],format='%B').dt.month
df_us_pivot.sort_values(by='months',inplace=True)
df_us_pivot.set_index('job_posted_month',inplace=True)
df_us_pivot.drop(columns='months')
```

```python
df_US_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')

df_US_software_pivot
```

```python
df_us_pivot_merge=df_us_pivot.merge(df_US_software_pivot, on='job_posted_month').drop(columns='months')
```

```python
top_5=df_us_pivot_merge.sum().sort_values(ascending=False).head()
top_5=top_5.index
```

```python
df_us_pivot_merge[top_5].plot(kind='line')
plt.title('Top 5 jobs')
plt.xlabel('2023')
plt.ylabel('job counts')
plt.ylim(0,20000)
plt.legend()
plt.show()
```
